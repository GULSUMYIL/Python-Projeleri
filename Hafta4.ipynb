```python
number=int(input("Please type a number: "))
while number>0:
  i=0
  while i<number:
    print(f"{i}",end=" ")
    i+=1
  print()
  number-=1
```

```python
sentence=input("Please type a sentence")
sentence=" "+sentence #ilk harfi görmesi için başına boşluk eklenir
sayac=0
while sayac<len(sentence)-1:
  if sentence[sayac]==" ":
    print(sentence[sayac+1])
    sayac +=1
```

```python
def print_many_times(text,times):
  while times!=0:
    times -=1
    print(text)
print_many_times("hi",3)
```

```python
def hash_square(length):
  for i in range(length):#pythonda iki tane döngüye gerek yok
    print("#"*length)

hash_square(5)
```

```python
def triangle(sayi):
  i=1
  while i<=sayi:
    print("#"*i)
    i +=1
triangle(3)
```

```python
def greatest_number(sayi1,sayi2,sayi3):
  if sayi1>sayi2 and sayi1>sayi3:
    return sayi1
  elif sayi2>sayi1 and sayi2>sayi3:
    return sayi2
  else:
    return sayi3

greatest_number(6,3,8)
```
